```python
def f(row):
    if flights['arr_delay'].isna() & flights['arr_time'].notna() & flights['sched_arr_time'].notna():
        val = flights['arr_time'] - flights['sched_arr_time']
    elif row['A'] > row['B']:
        val = 1
    else:
        val = -1
    return val
```

```python
flights['Date'] = flights['month'].astype(str) + flights['day'].astype(str) + flights['year'].astype(str)
flights['Date'] = pd.to_datetime(flights[['month', 'day', 'year']])

r= pd.to_datetime(flights[['year', 'month', 'day']])

flights['date'] = pd.to_datetime(flights[['day','month','year']])
flights['Day'] = flights['date'].dt.strftime('%A')
flights.head()
flights['Arr_delay?'] = np.where(flights['arr_delay'] > 0, 1, 0)
flights['Dept_delay?'] = np.where(flights['dep_delay'] > 0, 1, 0)




flights['Dept_late_and_reached_on_time'] = np.where(((flights['dep_delay'] > 0) & (flights['arr_delay'] <= 0)), 1, 0)

flights['Dept_late_and_arrived_late'] = np.where(((flights['dep_delay'] > 0) & (flights['arr_delay'] > 0)), 1, 0)
flights['Flights_departed_on_time_but_arrived_late'] = np.where(((flights['dep_delay'] <= 0) & (flights['arr_delay'] > 0)), 1, 0)
```

```python
Total_flights = flights.shape[0]
Flights_departed_late = sum(flights['Dept_delay?'])
Flights_arrived_late = sum(flights['Arr_delay?'])
Flights_arrived_late_because_depated_late = sum(flights['Dept_late_and_arrived_late'])
Flights_arrived_on_time_but_depated_late = sum(flights['Dept_late_and_reached_on_time'])
Flights_departed_on_time_but_arrived_late = sum(flights['Flights_departed_on_time_but_arrived_late'])

print('Total_flights - ', Total_flights)
print('Flights_departed_late - ', Flights_departed_late)
print('Flights_arrived_late - ', Flights_arrived_late)

print('Flights_arrived_on_time_but_departed_late - ', Flights_arrived_on_time_but_depated_late)
print('Flights_departed_on_time_but_arrived_late - ', Flights_departed_on_time_but_arrived_late)


print('Flights_arrived_late_because_depated_late - ', Flights_arrived_late_because_depated_late)

# % of flight delays due to flight departing late (because if a flight departed late then it is supposed to arrive late)

print( '% of flight delays due to flight departing late - ' + str(round(Flights_arrived_late_because_depated_late/ Flights_arrived_late*100,2))+'%')
print( '% of flight delays due to more flight time or flight arriving late - ' + str(round(Flights_departed_on_time_but_arrived_late/ Flights_arrived_late*100,2))+'%')

```
